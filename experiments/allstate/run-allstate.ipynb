# Allstate Claims Severity Dataset

```python
from experiment_runner import run_all
from params import gen_params_list
from util import average_results

data_h5 = "allstate-claims.h5"

df = pd.read_hdf(data_h5)
catfeats = list(range(0, 116))
df.iloc[:, catfeats] = df.iloc[:, catfeats].astype(np.int32) #.astype('category')
```

## Accurate parameter set

Note that these numbers do not correspond exactly to the numbers in the paper as they were generated by a different machine. The general conclusions are unchanged, however.

```python
params_list = gen_params_list({
    'method': ['bit', 'lgb'],
    "target": "loss",
    'learning_rate': 0.15,
    'categorical': [catfeats],
    'objective': 'huber', 'metric': 'mae_exp', 'niterations': [150], 'max_depth': [5],
    'feature_fraction': [0.8],
    'example_fraction': [1.0],
    'bit_sample_freq': [20],
    'bit_max_nbins': [16],
    'bit_discr_nbits': [4],
    'bit_sample_freq': [20],
    'bit_compr_threshold': [0.5]
}) + gen_params_list({
    'method': ['xgb', 'cat'],
    "target": "loss",
    'learning_rate': 0.15,
    'categorical': [catfeats],
    'objective': 'reg_l2', 'metric': 'mae_exp', 'niterations': [150], 'max_depth': [5],
    'feature_fraction': [0.8],
    'example_fraction': [1.0],
})
csv = run_all(df, params_list, nprocs=4, random_state=15)
```

```python
average_results(csv)
```
