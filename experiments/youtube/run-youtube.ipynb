# Youtube Trending Videos Dataset

```python
from experiment_runner import run_all
from params import gen_params_list
from util import average_results

data_h5 = "youtube.h5"

df = pd.read_hdf(data_h5)

def is_cat(iname):
    p1 = iname[1] in ['cat_id', 'country', 'cmtsdis', 'likedis', 'err']
    p2 = iname[1].startswith('txt_')
    return p1 | p2

catfeats = list(map(lambda x: x[0], filter(is_cat, enumerate(df.columns))))
df.iloc[:, catfeats] = df.iloc[:, catfeats].astype(np.int32) #.astype('category')
```

## Accurate parameter set

Note that these numbers do not correspond exactly to the numbers in the paper as they were generated by a different machine. The general conclusions are unchanged, however.

```python
params_list = gen_params_list({
    'method': ['bit'], #, 'cat'],
    "target": "viewslg",
    'categorical': [catfeats],
    'objective': 'huber', 'metric': 'mae',
    'niterations': [150], 'max_depth': [9],
    "learning_rate": 0.25,
    'feature_fraction': [1.0],
    'example_fraction': [1.0],
    'bit_discr_nbits': [2],
    'bit_sample_freq': [10],
    "bit_compr_threshold": [0.7], # 0.5 in paper
    "bit_max_nbins": 32,
}) + gen_params_list({
    'method': ['lgb', 'xgb', 'cat'], #, 'cat'],
    "target": "viewslg",
    'categorical': [catfeats],
    'objective': 'reg_l2', 'metric': 'mae',
    'niterations': [150], 'max_depth': [9],
    "learning_rate": 0.25,
    'feature_fraction': [1.0],
    'example_fraction': [1.0],
    "bit_max_nbins": 32,
})

csv = run_all(df, params_list, nprocs=4, random_state=1)
```

```python
average_results(csv)
```
